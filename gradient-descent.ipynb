```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def grad(x):
    return 2*x+5*np.cos(x)
def cost(x):
    return x**2+5*np.sin(x)
def MyGD1(eta,x0):
    x=[x0]
    y=[cost(x0)]
    for i in range(100):
        x_new=x[-1]-eta*grad(x[-1])
        if abs(x_new-x[-1])<=1e-3:
            break
        x.append(x_new)
        y.append(cost(x_new))
    return (x,y,i)
```

```python
data=[[],[]]
for i in range(-6000,6000,1):
    data[0].append(i/1000)
    data[1].append(cost(i/1000))
plt.plot(data[0],data[1],'r')
(x1,y1,it1)=MyGD1(0.01,-5)
(x2,y2,it2)=MyGD1(0.01,5)
plt.plot(x2,y2,'g')
plt.plot(x1,y1,'b.')
plt.show()
```
